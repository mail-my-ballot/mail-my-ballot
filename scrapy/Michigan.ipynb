# Michigan Data

```python
df_raw = pd.read_json('Michigan.jl', lines=True)
```

## County Data

```python
df_counties = df_raw[df_raw['type']=='county'].dropna(axis=1, how='all')
df_counties.sample(10, random_state=42)
```

### These two missing values are actually missing from the website

```python
df_counties[df_counties.isnull().any(axis=1)]
```

## Local Data

```python
df_local = df_raw[df_raw['type']=='local'].dropna(axis=1, how='all')
df_local.sample(10, random_state=42)
```

### These missing values appear to be missing from the website

```python
df_local[df_local['email'].isnull()]
```

## These are suffixes used for cities

```python
df_local['jurisdictionName'].str.split(' ').apply(lambda x: x[-1]).value_counts()
```
